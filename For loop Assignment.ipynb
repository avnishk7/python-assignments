## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
n = int(input("Check whether the number is prime or not: "))
count = 0
for i in range(2,n-1):
    if n % i == 0:
        count+= 1
        break
if count == 0:
    print("Its a Prime number")
else:
    print("Its Not a Prime number")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
# write your code here
n = int(input("Enter your number : "))
for i in range(2, n + 1):
    is_prime = True   # flag variable

    for j in range(2, i):
        if i % j == 0:
            is_prime = False
            break

    if is_prime:
         print(f"The number is prime: {i}")


```

## Question 20 (Hard)

**Title:** Perfect Number Check

### Problem Statement:
Take a number as input and check whether it is a perfect number or not.

### Approach:
Find sum of all factors using a loop and compare with the number.


```python
# write your code here
n = int(input("Check your number is perfect or not : "))
sum = 0
for i in range(1,n):
    if n % i == 0:
        sum = sum + i
if sum == n:
    print("The number is perfect")
else:
    print("The number is not perfect")
```
