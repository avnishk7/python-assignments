### Hard 2

**Question:**  
Find the second largest element.

**Approach:**  
Track largest and second largest values.


```python
# Write your code here
my_list = [12,22,55,60,67,4,23,5,39]
largest = my_list[0]
second_largest = my_list[0]
for i in my_list:
  if i > largest:
    second_largest = largest
    largest = i

  elif i > second_largest:
    second_largest = i
print(second_largest)
```
